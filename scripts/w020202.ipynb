```python
import seaborn as sns
import scipy.stats as st
def check_dist(x, fs=(20, 3), xlab="Dist", ylab="Freq", titlefont=20, axisfont=15, hspace=0.4, wspace=0.2):
    fig = plt.figure(figsize=fs)
    fig.tight_layout()
    
    ax1 = fig.add_subplot(1, 3, 1)
    ax1.set_title("Johnson SU", fontsize=titlefont)
    ax = sns.distplot(x, kde=False, fit=st.johnsonsu)
    ax.set(xlabel=xlab, ylabel=ylab)
    ax.set_xlabel(xlab, fontsize=axisfont)
    ax.set_ylabel(ylab, fontsize=axisfont)

    ax2 = fig.add_subplot(1, 3, 2, sharey=ax1)
    ax2.set_title("Log Normal", fontsize=titlefont)
    ax = sns.distplot(x, kde=False, fit=st.lognorm)
    ax.set(xlabel=xlab, ylabel=ylab)
    ax.set_xlabel(xlab, fontsize=axisfont)
    ax.set_ylabel(ylab, fontsize=axisfont)

    ax3 = fig.add_subplot(1, 3, 3, sharey=ax1)
    ax3.set_title("Normal", fontsize=titlefont)
    ax = sns.distplot(x, kde=False, fit=st.norm)
    ax.set(xlabel=xlab, ylabel=ylab)
    ax.set_xlabel(xlab, fontsize=axisfont)
    ax.set_ylabel(ylab, fontsize=axisfont)
    
    fig.subplots_adjust(wspace=wspace)
    plt.show()
```

```python
check_dist(raw.diffCSscore, xlab="measure of disease severity")
```

```python
# apply johnson transformation
def johnson_transform(x):
    gamma, eta, epsilon, lbda = st.johnsonsu.fit(x)
    yt = gamma + eta * np.arcsinh((x - epsilon) / lbda)
    return yt, gamma, eta, epsilon, lbda


# apply inverse of johnson transformation
def johnson_inverse(y, gamma, eta, epsilon, lbda):
    return lbda * np.sinh((y - gamma) / eta) + epsilon
```

```python
y_trans, gamma, eta, epsilon, lbda = johnson_transform(raw.diffCSscore)
check_dist(y_trans, xlab="measure of disease severity")
```

```python
y_orig = johnson_inverse(y_trans, gamma, eta, epsilon, lbda)
check_dist(y_orig, xlab="measure of disease severity")
```

```python
# normalize the whole dataframe (value type), and return the new dataframe, as well as the mean and standard deviation
# of each column in the old dataframe
def normalization(df):
    mean = df.mean()
    std = df.std()
    df = (df - mean) / std
    return df, mean, std


# normalize the whole dataframe (value type) using a certain mu and sigma
def normalization2(df, mu, sigma):
    df = (df - mu) / sigma
    return df
```

```python
yt_disease = X_meta_disease.y_trans
X_meta_disease = X_meta_disease.drop(['y_trans'], axis=1)

X_meta_disease, mu_meta_d, sigma_meta_d = normalization(X_meta_disease)
```

```python
from sklearn.model_selection import train_test_split
xtrain_meta_d, xval_meta_d, ytrain_meta_d, yval_meta_d = train_test_split(X_meta_disease, yt_disease, test_size = 0.2, random_state = 1)
```

## RF

### base-line model

```python
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor()
rf.fit(xtrain_meta_d, ytrain_meta_d)

# training set
rf_train_pred = pd.Series(rf.predict(xtrain_meta_d))
rf_train_pred_orig = johnson_inverse(rf_train_pred, gamma, eta, epsilon, lbda)
ytrain_meta_d_orig = johnson_inverse(ytrain_meta_d, gamma, eta, epsilon, lbda)
```

```python
# validation set
rf_val_pred = pd.Series(rf.predict(xval_meta_d))
rf_val_pred_orig = johnson_inverse(rf_val_pred, gamma, eta, epsilon, lbda)
yval_meta_d_orig = johnson_inverse(yval_meta_d, gamma, eta, epsilon, lbda)
```

```python
from sklearn.metrics import mean_squared_error
mse_rf_train = mean_squared_error(rf_train_pred_orig, ytrain_meta_d_orig)
mse_rf_val = mean_squared_error(rf_val_pred_orig, yval_meta_d_orig)
np.sqrt(mse_rf_train), np.sqrt(mse_rf_val)
```

It seems that the base-line rf model results in overfitting. 

### random search

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
param_grid = {
    'max_depth': [7,10,13,15,17],
    'max_features': [5,7,10],
    'min_samples_split': [2,5,10],
    'n_estimators': [20,50,100]
}
rf_random_meta_d = RandomizedSearchCV(estimator = rf, param_distributions = param_grid, n_iter=100, cv = 5)
```
